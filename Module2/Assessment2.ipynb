```python
# hotel
import random

class HotelReservation:
    def __init__(self):
        self.name = ''
        self.contact = ''
        self.room_type = ''
        self.nights = 0
        self.booking_id = None
        self.active_booking = False
        self.room_prices = {
            "Standard": 2000,
            "Deluxe": 3500,
            "Suite": 5000
        }
        self.menu()

    def menu(self):
        user_input = input('''
        Welcome to Hotel Reservation System
        1. Book a Room
        2. View Booking
        3. Cancel Booking
        4. Exit
        ''')

        if user_input == '1':
            if self.active_booking:
                print("You already have an active booking!")
                self.menu()
            else:
                self.book_room()
        elif user_input == '2':
            self.view_booking()
        elif user_input == '3':
            self.cancel_booking()
        elif user_input == '4':
            print("Exiting the system... Thank you for using Hotel Reservation!")
            exit()
        else:
            print("Invalid choice! Please try again.")
            self.menu()

    def book_room(self):
        self.name = input("Enter your name: ")
        self.contact = input("Enter your 10-digit contact number: ")

        if len(self.contact) != 10 or not self.contact.isdigit():
            print("Invalid contact number! Please try again.")
            self.book_room()
            return

        print("\nAvailable room types:")
        for room, price in self.room_prices.items():
            print(f"{room} - ₹{price}/night")

        choice = input("Enter room type (Standard/Deluxe/Suite): ")
        if choice not in self.room_prices:
            print("Invalid room type! Please try again.")
            self.book_room()
            return

        self.room_type = choice
        self.nights = int(input("Enter number of nights: "))
        total_cost = self.room_prices[self.room_type] * self.nights
        print(f"Total cost: ₹{total_cost}")

        confirm = input("Confirm booking? (yes/no): ")
        if confirm.lower() == 'yes':
            self.booking_id = random.randint(1000, 9999)
            self.active_booking = True
            print(f"Booking confirmed! Your booking ID is {self.booking_id}")
        else:
            print("Booking cancelled.")
        
        self.menu()

    def view_booking(self):
        if self.active_booking:
            print("\n--- Booking Details ---")
            print(f"Name: {self.name}")
            print(f"Contact: {self.contact}")
            print(f"Room Type: {self.room_type}")
            print(f"Nights: {self.nights}")
            print(f"Booking ID: {self.booking_id}")
        else:
            print("No active booking found.")
        self.menu()

    def cancel_booking(self):
        if self.active_booking:
            confirm = input("Are you sure you want to cancel booking? (yes/no): ")
            if confirm.lower() == 'yes':
                self.active_booking = False
                self.booking_id = None
                print("Booking cancelled successfully.")
            else:
                print("Cancellation aborted.")
        else:
            print("No booking to cancel.")
        self.menu()



```
